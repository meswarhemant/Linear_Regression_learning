```python
# Take "Kills" as the response variable and "Ln Match Kill Rank" as the predictor. 
# Perform a simple linear regression.
import statsmodels.api as sm


#Define the dependent variable (target)
y = df['Kills']

# Define the independent variables (features)
X = df['Ln Match Kill Rank']
X = sm.add_constant(X)  # Adds a constant (intercept) term
#y = target_column

model1 = sm.OLS(y, X).fit()
print(model1.summary())


```

```python
# Show the plot for the above simple linear regression model.

import matplotlib.pyplot as plt

plt.scatter(df['Ln Match Kill Rank'], y, label='Original Data')
plt.plot(df['Ln Match Kill Rank'], model1.predict(X), color='red', label='OLS Regression')
plt.xlabel('Ln Match Kill Rank')
plt.ylabel('Kills')
plt.legend()
plt.show()
```

```python
# Using 'Winplaceperc' as the response variable, perform a full linear regression model
import statsmodels.api as sm


# Define the response variable (dependent variable)
y = df['Winplaceperc']

# Select all columns except 'Winplaceperc' as the predictor variables (independent variables)
X = sm.add_constant(df.drop('Winplaceperc', axis=1))

# Fit the full linear regression model
model2 = sm.OLS(y, X).fit()

# Print the model summary
print(model2.summary())
```

```python
# Perform a 4-feature multiple linear regression using Winplaceperc as the response variable
# and "Kills", "Walk distance", "Matchkillrank", and "Boosts" as the predictors.



# Define the predictor variables (independent variables)
X = df[['Boosts', 'Walkdistance','Matchkillrank','Kills',]]

# Add a constant term to the predictor variables for the intercept
X = sm.add_constant(X)

# Define the response variable (dependent variable)
y = df['Winplaceperc']

# Fit the multiple linear regression model
model3 = sm.OLS(y, X).fit()

# Print the model summary
print(model3.summary())
```

```python
# Display the partial regression plots on the 4-feature multiple linear regression


fig = sm.graphics.plot_partregress_grid(model3)
fig.tight_layout(pad=1.0)
```

```python
# Draw the Q-Q plot for the 4-feature multiple linear regression


#Running plot & giving it a title
from scipy import stats
stats.probplot(model3.resid, dist="norm", plot= plt)
plt.title("Model Residuals Q-Q Plot")
```
